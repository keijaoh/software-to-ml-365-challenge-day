```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
 
 
# Authenticate and create the PyDrive client.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```python
import pandas as pd
```

```python
file_link = "FileLocationInGoogleDrive"
file_name = "training_set.csv"


# to get the id part of the file
id = file_link.split("/")[-2]
 
downloaded = drive.CreateFile({'id':id})
downloaded.GetContentFile(file_name) 
 
hema_csv = pd.read_csv(file_name)
```

```python
#this command is to show the plots inside of jupyter notebook
#%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
#prepare the data
x_axis = ['Jan','Feb','Mar','Apr']
y_axis = [12,24,36,66]

plt.plot(x_axis,y_axis)

#add label
plt.xlabel('Months')
plt.ylabel('Number of Sales');
```

```python
#reccomended approach
#object oriented API
#figsize = height, width
figure, axis = plt.subplots(figsize= (5,5))

new_y_axis = [150,200,120,250]

#plot the data
axis.plot(x_axis,new_y_axis);

#customize the plot 

axis.set(
    title= 'Sales This Year',
    xlabel= 'Months',
    ylabel= 'Number of Sales'
);

#save the plot
#figure.savefig('images/plots/destination.png')
```
